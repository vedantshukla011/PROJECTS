```python
from selenium import webdriver
```

```python
from selenium.webdriver.common.by import By
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Edge(executable_path = 'C:\\Users\\hp\\Desktop\\msedgedriver')
driver
```

```python
driver.get('http://kanview.ks.gov/PayRates/PayRates_Agency.aspx')
```

```python
c1=driver.find_element(By.LINK_TEXT,"Adjutant General")
c1.click()
```

```python
b=driver.find_element(By.ID,"MainContent_uxLevel2_JobTitles_uxJobTitleBtn_0")
b
```

```python
from selenium.common.exceptions import NoSuchElementException

dataFrameList = []

i = 0
while True : 
    try : 
        l= driver.find_element(By.ID,'MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
        l.click()

        data = BeautifulSoup(driver.page_source, 'lxml')
        emp_table = data.table
        df = pd.read_html(str(emp_table), header = 0)
        dataFrameList.append(df[0])
        driver.back()
        i += 1
    except NoSuchElementException :
        break
```

```python
dataFrameList
```

```python
df = pd.concat(dataFrameList, ignore_index = True)
df
```
